```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("Asociaciones.xlsx", header=None, skiprows=1)
df.head(50)
```

```python
plants = df[0].dropna()
opt = ['Favorable', np.nan]
rel = df.loc[~df[1].isin(opt),1:3].reset_index(drop=True)
def get_splits(plant):
    rest, family = plant.rsplit(' ', 1)
    family = family.capitalize()
    rest = rest.split()
    name = ' '.join([w for w in rest if w.isupper()]).capitalize()
    csname = ' '.join([w for w in rest if not w.isupper()]).capitalize()
    return name, csname, family
plant = pd.DataFrame()
plant['Name'], plant['scientific'], plant['family'] = \
    zip(*plants.map(get_splits))
rel = rel.rename(columns={1: "Favorable", 2: "Desfavorable", 3: 'Descripción'})
pd.concat([plant, rel], axis=1)
```
